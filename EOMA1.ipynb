````python
def get_overtime_pay(monthly_fee,hours_rendered):
    
    hourly_rate = monthly_fee / 20
    
    if hours_rendered > 20:
        overtime_pay = ((hours_rendered - 20) * hourly_rate) * 2
    
    else:
        overtime_pay = 0
        
    return overtime_pay
````

### 3.4: Bank Credit Advise



**B. Bank Credit Advise**

Generate a credit advise file each for BPI and BDO covering interns and consultants. The files must be named **./output/su-bpi-payment-advise.csv** and **./output/su-bdo-payment-advise.csv** respectively and should look like the following:

```
account_number,amount
109350005001,4800.0
109350005002,3275.0
...
109350005003,3375.0
...
109350006001,9000
```



````python
bpi_advise = []
bdo_advise = []

for interns in interns_payroll:
    for intern in interns_master_dict:
        
        if interns['id'] == intern:
            intern_advise = {'account_number': interns_master_dict[intern]['account_number'],
                             'amount':interns['takehome_pay']}

            if interns_master_dict[intern]['bank'] == "BPI":
                bpi_advise.append(intern_advise)
            
            elif interns_master_dict[intern]['bank'] == "BDO":
                bdo_advise.append(intern_advise)  
                
for consultants in consultants_payroll:
    for consultant in consultants_master_dict:
        
        if consultants['tin'] == consultant:
            consultant_advise = {'account_number': consultants_master_dict[consultant]['account_number'],
                                 'amount':consultants['net_pay']}
            
            if consultants_master_dict[consultant]['bank'] == "BPI":
                bpi_advise.append(consultant_advise)
            
            elif consultants_master_dict[consultant]['bank'] == "BDO":
                bdo_advise.append(consultant_advise)

with open("./output/su-bpi-payment-advise.csv", "w", newline="") as csvfile:
    field_names = ['account_number', 'amount']
    writer = csv.DictWriter(csvfile, fieldnames=field_names)
    writer.writeheader()
    
    for r in bpi_advise:
        writer.writerow(r)
        
with open("./output/su-bdo-payment-advise.csv", "w", newline="") as csvfile:
    field_names = ['account_number', 'amount']
    writer = csv.DictWriter(csvfile, fieldnames=field_names)
    writer.writeheader()
    
    for r in bdo_advise:
        writer.writerow(r)
````
